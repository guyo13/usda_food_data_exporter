```python
import export_food_data
import pandas as pd
```

```python
food_nutrients = export_food_data.get_nutri_data_table()
```

```python
food_nutrients
```

```python
food_nutrients.sort_values("fdc_id", 
                           inplace=True,ignore_index=True)
```

```python
df = food_nutrients.pivot(index="fdc_id",columns="nutrient_id",values="amount")
```

```python
df
```

```python
foods = export_food_data.get_food_table(usecols=["description", "data_type"])
```

```python
foods
```

```python
final = pd.concat([df, foods], axis=1, join='inner')
```

```python
final
```

```python
nutri_dict= export_food_data.get_nutrient_dict()
```

```python
nutri_dict
```

```python
final.rename(inplace=True, axis=1, mapper= lambda x: nutri_dict[x]["name"] if isinstance(x, int) else x)
```
